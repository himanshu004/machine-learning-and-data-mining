```python
# translation

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

size = int(input('Enter the number of coordinates:'))
mat = np.ones(4 * size,dtype = 'int').reshape(size,4)

print('Enter the co-ordinates:')
for i in range(0,size):
        x,y,z = (map(int,input().split()))
        mat[i][0] = x
        mat[i][1] = y
        mat[i][2] = z

            
print('The matrix you entered is:')
print(mat)

x = mat[:,0]
new_x = np.append(x,x[0])
y = mat[:,1]
new_y = np.append(y,y[0])
z = mat[:,2]
new_z = np.append(z,z[0])

plt3D = plt.axes(projection = '3d')
plt3D.plot3D(new_x,new_y,new_z)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
    
tx = int(input('Enter tx:'))
ty = int(input('Enter ty:'))

t = np.array([[1,0,1],[0,1,0],[tx,ty,1]])
new_mat = mat.dot(t)

print('The matrix after translation is:')
print(new_mat)

x = new_mat[:,0]
new_x = np.append(x,x[0])
y = new_mat[:,1]
new_y = np.append(y,y[0])

plt.plot(new_x,new_y,'b*--',markersize = 20,linewidth = 3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
