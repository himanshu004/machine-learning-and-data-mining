```python
# translation

import numpy as np
import matplotlib.pyplot as plt

size = int(input('Enter the number of coordinates:'))
mat = np.ones(3 * size,dtype = 'int').reshape(size,3)

print('Enter the co-ordinates:')
for i in range(0,size):
        x,y = (map(int,input().split()))
        mat[i][0] = x
        mat[i][1] = y
        
print('The matrix you entered is:')
print(mat)

x = mat[:,0]
new_x = np.append(x,x[0])
y = mat[:,1]
new_y = np.append(y,y[0])

plt.plot(new_x,new_y,'r*--',markersize = 20,linewidth = 3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
    
tx = int(input('Enter tx:'))
ty = int(input('Enter ty:'))

t = np.array([[1,0,1],[0,1,0],[tx,ty,1]])
new_mat = mat.dot(t)

print('The matrix after translation is:')
print(new_mat)

x = new_mat[:,0]
new_x = np.append(x,x[0])
y = new_mat[:,1]
new_y = np.append(y,y[0])

plt.plot(new_x,new_y,'r*--',markersize = 20,linewidth = 3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
# scaling

import numpy as np
import matplotlib.pyplot as plt

size = int(input('Enter the number of coordinates:'))
mat = np.zeros(3 * size,dtype = 'int').reshape(size,3)

print('Enter the co-ordinates:')
for i in range(0,size):
        x,y = (map(int,input().split()))
        mat[i][0] = x
        mat[i][1] = y
        
print('The matrix you entered is:')
print(mat)

x = mat[:,0]
new_x = np.append(x,x[0])
y = mat[:,1]
new_y = np.append(y,y[0])

plt.plot(new_x,new_y,'r*--',markersize = 20,linewidth = 3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
    
sx = int(input('Enter Sx:'))
sy = int(input('Enter Sy:'))

t = np.array([[sx,0],[0,sy],[0,0]])
new_mat = mat.dot(t)

print('The matrix after scaling is:')
print(new_mat)

x = new_mat[:,0]
new_x = np.append(x,x[0])
y = new_mat[:,1]
new_y = np.append(y,y[0])

plt.plot(new_x,new_y,'r*--',markersize = 20,linewidth = 3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
